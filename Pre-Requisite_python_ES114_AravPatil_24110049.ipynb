```python
#Question 1
natural_nos=[]
squared=[]
sum_even=0
prod_odd=1
count_nos=0
count=0
dev_top=0
median=0
for i in range(1,21):
    natural_nos.append(i)
for e in natural_nos:
    count_nos+=1
    count+=e
    if e%2==0:
        sum_even+=e
    else:
        prod_odd*=e
if count%2==0:
    l_num=natural_nos[count_nos//2-1]
    u_num=natural_nos[(count_nos//2)]
    median=(l_num+u_num)/2
else:
    median=natural_nos[count_nos//2-1]
mean=count/count_nos
for b in natural_nos:
    temp = (mean-b)**2
    squared.append(temp)
for g in squared:
    dev_top+=g
std_dev = (dev_top/count_nos)**(1/2)
print("Sum:",sum_even)
print("Product:",prod_odd)
print("Mean:",mean)
print("Median:",median)
print("Standard Deviation:",std_dev)
```

```python
#Question 3
import matplotlib.pyplot as plt
def fibb(n):
    ls=[0,1]
    if n==0:
        return [0]
    elif n==1:
        return [0,1]
    for i in range(2,n):
        temp=ls[-1]+ls[-2]
        ls.append(temp)
    return ls
n = int(input(""))
nos=[]
for i in range(0,n):
    nos.append(i)
print(fibb(n))
plt.title('Number v/s Corresponding Fibonnaci number')
plt.xlabel('Numbers')
plt.ylabel('Fibonacci Number')
plt.plot(nos,fibb(n))
```

```python
#Question 15
def find_anagrams(word, words):
    sorted_word = ''.join(sorted(word))
    anagrams = [w for w in words if ''.join(sorted(w)) == sorted_word]
    return anagrams
word = "listen"
words = ["enlist", "google", "inlets", "banana"]
result = find_anagrams(word, words)
print(result)
```

```python
#Question 16
ls1=[]
ls2=[]
inter=[]
diff=[]
uni=[]
number1 = int(input("Number of elements in list1:"))
number2 = int(input('Number of elements in list2:'))
while number1>=1:
    temp=int(input('Enter the number for list1:'))
    ls1.append(temp)
    number1-=1
while number2>=1:
    temp=int(input('Enter the number for list2:'))
    ls2.append(temp)
    number2-=1
for num1 in ls1:
    if num1 in ls2:
        uni.append(num1)
        inter.append(num1)
    else:
        diff.append(num1)
        uni.append(num1)
for num2 in ls2:
    if num2 not in ls1:
        diff.append(num2)
        uni.append(num2)
print("Union:",uni)
print("Intersection:",inter)
print("Symmetric difference:",diff)
```

```python
#Question 18
import matplotlib.pyplot as plt

def read_and_analyze_file(filename):
    try:
        with open(filename, 'r') as file:
            text = file.read()
    except FileNotFoundError:
        print(f"File '{filename}' not found.")
        return
    text = text.lower()
    cleaned_text = ""
    for char in text:
        if char.isalnum() or char.isspace():
            cleaned_text += char
    words = cleaned_text.split()
    total_words = len(words)
    print(f"Total number of words: {total_words}")
    word_freq = {}
    for word in words:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1
    most_common_word = max(word_freq, key=word_freq.get)
    highest_frequency = word_freq[most_common_word]
    print(f"Most frequent word: '{most_common_word}' (appears {highest_frequency} times)")
    visualize_word_frequencies(word_freq)
def visualize_word_frequencies(word_freq):
    # Get the 10 most common words
    sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:10]
    words = [item[0] for item in sorted_words]
    frequencies = [item[1] for item in sorted_words]
    plt.figure(figsize=(10, 6))
    plt.bar(words, frequencies, color='skyblue', edgecolor='black')
    plt.title('Top 10 Most Frequent Words', fontsize=16)
    plt.xlabel('Words', fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()
    plt.show()
if __name__ == "__main__":
    filename = input("Enter the filename: ")
    read_and_analyze_file(filename)
```

```python
#Question 19
import random
import matplotlib.pyplot as plt
fileh=open('temp.txt','r+')
n=100
ls=[]
count=0
sum=0
values=[]
while n>=1:
    temp=random.randint(1,50)
    ls.append(temp)
    n=n-1
for i in ls:
    i=str(i)
    fileh.write(i)
    fileh.write('\n')
fileh.seek(0)
a = fileh.readlines()
for v in a:
    v=v.strip()
    values.append(v)
values.remove('')
plt.hist(values, bins=10, color='blue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Random Numbers')
plt.xlabel('Number Range')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
for h in values:
    h=int(h)
    count+=1
    sum+=h
print('Mean:',sum/count)
fileh.close()
```

```python
#Question 21
import matplotlib.pyplot as plt
n = int(input("How many numbers in the list?:"))
ls=[]
x_axis=[]
while n>=1:
    x_axis.append(n)
    temp=float(input("Enter the number:"))
    ls.append(temp)
    n-=1
ls.sort()
x_axis.sort()
count=len(ls)
sum_ls=0
freq={}
mode=0
elements_with_max_freq=[]
max_freq=0
for i in ls:
    sum_ls+=i
    if i in freq:
        freq[i]+=1
    else:
        freq[i]=1
for v in freq.values():
    if v>max_freq:
        max_freq=v
for key,value in freq.items():
    if value==max_freq:
        elements_with_max_freq.append(key)
if count%2==0:
    l_num=ls[count//2-1]
    u_num=ls[count//2]
    median=(l_num+u_num)/2
else:
    median=ls[count//2]
mean=sum_ls/count
sd_ls=[]
for i in ls:
    a=(i-mean)**2
    sd_ls.append(a)
sum_sd_ls=0
for b in sd_ls:
    sum_sd_ls+=b
sd=((sum_sd_ls)/count)**(1/2)
print("Mean:",mean)
print("Median:",median)
print("Mode:",elements_with_max_freq)
print("Standard deviation:",sd)
plt.boxplot(ls)
```
